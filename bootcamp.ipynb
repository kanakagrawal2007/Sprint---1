```python
import pandas as pd
data={
    "student_id":[101,102,103,104,105,106],
    "attendance_percent":[92,67,81,45,74,88],
    "assignment_score":[18,12,15,8,14,19],
    "quiz_score":[72,48,65,30,55,80],
    "lab_completed":[True,False,True,False,True,True]
}
df=pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['total_score']=df['assignment_score']+df['quiz_score']
df
```

```python
df['eligible']=(
    (df['attendance_percent']>=75) &
    (df['total_score']>=70) &
    (df['lab_completed'])
)
df
```

```python
eligible_students=df[df['eligible']==True]
eligible_students
```

```python
import matplotlib.pyplot as plt
```

```python
# attendence per student
plt.bar(df['student_id'], df['attendance_percent'], color='skyblue')
plt.xlabel('Student ID')
plt.ylabel('Attendance (%)')
plt.title('1. Attendance Percentage by Student')
plt.grid(True)
plt.show()
```

```python
# attendence vs quiz score
plt.scatter(df['attendance_percent'], df['quiz_score'], color='red',s=100)
plt.xlabel('Attendance %')
plt.ylabel('Quiz Score')
plt.title('2. Correlation: Attendance vs Quiz Score')
plt.grid(True)
plt.show()
```

```python
 # Lab Completion Ratio
lab_counts = df['lab_completed'].value_counts()
plt.pie(lab_counts, labels=['Completed', 'Incomplete'], autopct='%1.1f%%', colors=['green', 'orange'])
plt.title('3. Lab Completion Status')
plt.show()
```

```python
# Quiz Score Trend
plt.plot(df['student_id'], df['quiz_score'], marker='o', linestyle='-', color='purple')
plt.xlabel('Student ID')
plt.ylabel('Quiz Score')
plt.title('4. Quiz Scores Across Students')
plt.grid(True)
plt.show()
```

```python
# Distribution of Assignment Scores
plt.hist(df['assignment_score'], bins=5, color='gold', edgecolor='black')
plt.xlabel('Assignment Score')
plt.ylabel('Frequency')
plt.title('5. Distribution of Assignment Scores')
plt.show()
```

```python
# Spread of Quiz Scores
plt.boxplot(df['quiz_score'])
plt.title('7. Statistical Spread of Quiz Scores')
plt.ylabel('Scores')
plt.show()
```

```python
# Assignment Scores
plt.barh(df['student_id'], df['assignment_score'], color='salmon')
plt.xlabel('Assignment Score')
plt.ylabel('Student ID')
plt.title('8. Assignment Scores (Horizontal)')
plt.show()
```

```python
#  Cumulative Performance Look
plt.stackplot(df['student_id'], df['attendance_percent'], labels=['Attendance'])
plt.title('9. Attendance Area Chart')
plt.ylabel('Percentage')
plt.show()
```

```python
# Average Quiz Score by Lab Completion
avg_scores = df.groupby('lab_completed')['quiz_score'].mean()
avg_scores.plot(kind='bar', color=['tomato', 'lightgreen'])
plt.xticks([0, 1], ['No Lab', 'Lab Done'], rotation=0)
plt.ylabel('Average Quiz Score')
plt.title('10. Impact of Lab Completion on Quiz Scores')
plt.show()
```

```python
#Find the average final_project_score.
#•	Find students with study_hours_per_week greater than 8.
#•	Compare average marks of Mumbai students and Pune students.
#•	Create a new column called performance_level.
#•	Save only selected columns to a new CSV.
```

```python
stu=df[df["study_hours_per_week"]>8]
print(stu[["name","study_hours_per_week"]])
```

```python
df["average_marks"] = df[["math", "science", "english"]].mean(axis=1)

mumbai_avg = df[df["city"] == "Mumbai"]["average_marks"].mean()
pune_avg = df[df["city"] == "Pune"]["average_marks"].mean()
print(mumbai_avg)
print(pune_avg)
```

```python
def performance_level(avg):
    if avg >= 85:
        return "Excellent"
    elif avg >= 70:
        return "Good"
    elif avg >= 50:
        return "Average"
    else:
        return "Poor"

df["performance_level"] = df["average_marks"].apply(performance_level)
print(df[["name","performance_level"]])
```

```python
selected_cols = df[["student_id", "name", "city", "average_marks", "performance_level", "final_project_score"]]
selected_cols.to_csv("student_performance_summary.csv", index=False)
```
